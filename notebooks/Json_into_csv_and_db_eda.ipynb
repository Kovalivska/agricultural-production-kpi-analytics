```python
import ast

# Liste zur Speicherung der extrahierten Dropdown-Optionen
dropdown_options_list = []

if 'df_inputs_full' in locals():
    # Untersuche die 'dropdownOptions'-Spalte in df_inputs_full
    print("Beispielwerte aus der 'dropdownOptions'-Spalte:")
    display(df_inputs_full['dropdownOptions'].head())

    print("\nEinzigartige Werte in der 'dropdownOptions'-Spalte:")
    unique_dropdown_options = df_inputs_full['dropdownOptions'].unique()
    if len(unique_dropdown_options) > 20:
        print(f"Es wurden {len(unique_dropdown_options)} einzigartige Werte gefunden. Zeige die Häufigkeiten der häufigsten an:")
        display(df_inputs_full['dropdownOptions'].value_counts().head())
    else:
        display(unique_dropdown_options)

    print("\nDatentyp der 'dropdownOptions'-Spalte:")
    print(df_inputs_full['dropdownOptions'].dtype)

    # Überprüfe auf Nicht-String- oder Nicht-Null-Werte, wenn das erwartete Format String ist
    print("\nÜberprüfung auf Nicht-String-/Nicht-Null-Werte in 'dropdownOptions':")
    non_string_non_null = df_inputs_full[df_inputs_full['dropdownOptions'].apply(lambda x: not isinstance(x, str) and pd.notnull(x))]
    if not non_string_non_null.empty:
        print("Nicht-String-/Nicht-Null-Werte gefunden:")
        display(non_string_non_null)
    else:
        print("Alle Nicht-Null-Werte sind vom Typ String.")


    # Iteriere durch jede Zeile in df_inputs_full
    for index, row in df_inputs_full.iterrows():
        input_id = row['id'] # Hole die ID des Eingabefeldes
        dropdown_options_str = row['dropdownOptions']

        # Werte die String-Darstellung der Liste sicher aus
        try:
            # Überprüfe, ob der String nicht leer ist und eine gültige Listendarstellung ist
            if isinstance(dropdown_options_str, str) and dropdown_options_str.strip():
                options_list = ast.literal_eval(dropdown_options_str)

                # Überprüfe, ob das ausgewertete Ergebnis eine Liste ist
                if isinstance(options_list, list):
                    for option in options_list:
                        # Stelle sicher, dass jedes Element in der Liste ein Dictionary ist und 'value' und 'label' enthält
                        if isinstance(option, dict) and 'value' in option and 'label' in option:
                            dropdown_options_list.append({
                                'input_id': input_id, # Verknüpfung zurück zur inputs_full Tabelle
                                'option_value': option['value'],
                                'option_label': option['label']
                            })
                        else:
                            # print(f"Warnung: Unerwartetes Format für eine Option in Zeile {index}: {option}")
                            pass # Warnung auskommentiert, um die Ausgabe übersichtlich zu halten
                else:
                    # print(f"Warnung: 'dropdownOptions' in Zeile {index} ist nach der Auswertung keine Liste: {options_list}")
                    pass # Warnung auskommentiert
            # Wenn der String leer oder NaN ist, bedeutet das, dass keine Dropdown-Optionen vorhanden sind, was in Ordnung ist.
        except (ValueError, SyntaxError) as e:
            print(f"Fehler beim Auswerten von dropdownOptions in Zeile {index}: {dropdown_options_str} - {e}")
        except Exception as e:
            print(f"Ein unerwarteter Fehler ist in Zeile {index} aufgetreten: {dropdown_options_str} - {e}")


    # Erstelle einen neuen DataFrame aus den extrahierten Optionen
    df_dropdown_options = pd.DataFrame(dropdown_options_list)

    print("\nDataFrame mit extrahierten Dropdown-Optionen erstellt.")
    display(df_dropdown_options.head())
    print(f"Form von df_dropdown_options: {df_dropdown_options.shape}")
else:
    print("Fehler: DataFrame 'df_inputs_full' nicht gefunden. Bitte stellen Sie sicher, dass die Daten aus den vollen CSVs geladen wurden.")
```
